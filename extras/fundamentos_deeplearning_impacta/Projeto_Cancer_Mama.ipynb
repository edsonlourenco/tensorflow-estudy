# Etapa 1: Importação das Bibliotecas

```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
```

# Etapa 2: Improtação de base dados

```python
# exames de cancer: https://drive.google.com/file/d/10dkV37k_vLeGrcBLM8Cc4som35HnOPF7/view
# saidas da analise: https://drive.google.com/file/d/1T1CMGwFfso5iZ2zfqQdGuafuQvcJnxF2/view
entradas = pd.read_csv('entradas_breast.csv')
saidas = pd.read_csv('saidas_breast.csv')
```

# Etapa 4: Tratamento de base de dados

```python
from sklearn.model_selection import train_test_split
```

```python
previsores_treinamento, previsores_teste, classe_treinamento, classe_test = train_test_split(entradas, saidas, test_size=0.25)
```

```python
previsores_treinamento
```

```python
previsores_teste
```

# Etapa 5: Construção e criação do modelo

```python
model = keras.Sequential()
```

```python
model
```

```python
model.add(keras.layers.Dense(units = 16, activation = 'relu', input_dim = 30, use_bias = True))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(units = 16, activation = 'relu', use_bias = True))
model.add(keras.layers.Dropout(0.5))

model.add(keras.layers.Dense(units = 1, activation = 'sigmoid', use_bias = True)) # problemas binários entre 1 ou 0
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['binary_accuracy'])
#model.compile(optimizer = keras.optimizers.Adam(learning_rate=0.1), loss = keras.losses.BinaryCrossentropy(), metrics = ['binary_accuracy'])
```

```python
model.fit(previsores_treinamento, classe_treinamento, batch_size=10, epochs = 1000, verbose=True)
```

# Etapa 6: Avaliação do Modelo

```python
resutlado = model.evaluate(previsores_teste, classe_test)
```

```python
resutlado
```

# Etapa 7: Salvando e produtizando o modelo

```python
model_json = model.to_json()
```

```python
model_json
```
