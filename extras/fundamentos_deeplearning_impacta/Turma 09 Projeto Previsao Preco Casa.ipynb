```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
house_df = pd.read_csv('kc-house-data.csv')
```

```python
house_df.info()
```

```python
sns.scatterplot(x = 'sqft_living', y = 'price', data = house_df);
```

```python
f, ax = plt.subplots(figsize = (20,20))
sns.heatmap(house_df.corr(), annot = True)
```

```python
house_df.hist(bins = 20, figsize= (20,20), color = 'b');
```

```python
selected_features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above','sqft_basement']
```

```python
X = house_df[selected_features]
```

```python
X
```

```python
y = house_df['price']
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
y.shape
```

```python
y = y.values.reshape(-1,1)
```

```python
y
```

```python
y_scaled = scaler.fit_transform(y)
```

```python
y_scaled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.30)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
model = tf.keras.models.Sequential()
```

```python
model.add(tf.keras.layers.Dense(units = 4, activation = 'relu' , input_shape=(7,)))
model.add(tf.keras.layers.Dense(units = 4, activation = 'relu' ))
model.add(tf.keras.layers.Dense(units = 1, activation = 'linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
epochs_hist = model.fit(X_train, y_train, epochs = 25, validation_split = 0.2)
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title( 'Modelo Loss Progress Durante Treino')
plt.xlabel('Epochs')
plt.ylabel('Training and Validation Loss')
plt.legend('Training Loss', 'Validation Loss');
```
